```python
#extract (x,y) data for plotting
def str2plot(s):
    p_qubit=[]
    p_block=[]
    s=s.split()
    n=len(s)//2
    print(n)
    for i in range(n):
        p=s[i*2]
        q=s[i*2+1]
        p=p.replace('{','').replace('}','').replace(',','').replace(';','')
        q=q.replace('{','').replace('}','').replace(',','').replace(';','')
        p=float(p)
        q=float(q)
        p_qubit.append(p)
        p_block.append(q)

#    print(p_qubit,p_block)
    return p_qubit,p_block
#p_qubit,p_block = data2plot(data['data_map'])

s='{{0.01, 9.99999999995449e-6}, {0.02, 7.999999999996898e-5}, {0.03, 0.00043999999999999595}, {0.04, 0.0018000000000000238}, {0.05, 0.0041400000000000325}, {0.06, 0.009639999999999982}, {0.07, 0.017329999999999957}, {0.08, 0.031710000000000016}, {0.09, 0.05069999999999997}, {0.1, 0.07230999999999999}, {0.11, 0.10211999999999999}, {0.12, 0.13466}, {0.13, 0.17276999999999998}, {0.14, 0.20903000000000005}};'
p_qubit,p_block = str2plot(s)
```
